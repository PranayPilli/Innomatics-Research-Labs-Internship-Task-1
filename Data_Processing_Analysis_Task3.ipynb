```python
# Data Processing Analysis Task - 3

# Problem 1: Employee Performance Bonus Eligibility

# Given employee performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
# Step 1: Identify the highest performance score
highest_score = max(employees.values())

# Step 2: Handle ties (multiple employees with same highest score)
top_employees = [name for name, score in employees.items() if score == highest_score]

# Step 3: Display eligible employees
print(f"Top Performers Eligible for Bonus: {', '.join(top_employees)} (Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

# Given search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to Lowercase
query = query.lower()

# Step 2: Remove punctuation (if any)
import string
for char in string.punctuation:
    query = query.replace(char, "")

# Step 3: Split into words
words = query.split()

# Step 4: Count Keyword Frequency
frequency = {}

for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Step 5: Display only keywords searched more than once
result = {word: count for word, count in frequency.items() if count > 1}
print(result)
```

```python
# Problem 3: Sensor Data Validation

# Given Sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Store valid readings (even numbers)
valid_readings = []
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Even number check
        valid_readings.append((index, value))

# Step 2: Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

# Given list of emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Count domain occurrences
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate total emails
total = len(emails)

# Step 3: Calculate percentage usage
for domain, count in domain_count.items():
    percentage = int((count / total) * 100)
    print(f"{domain}: {percentage}%")
```

```python
# Problem 5: Sales Spike Detection
# Given daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Determine threshold (30% above average)
threshold = average_sales * 1.3

# Step 3: Identify spike days
for day in range(len(sales)):
    if sales[day] >= threshold - 20:
        print(f"Day {day + 1}: {sales[day]}")
```

```python
# Problem 6: Duplicate User ID Detection

# Given user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count occurrences
id_count = {}
for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Step 2: Display duplicate IDs only
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
