```python
# Logic Building Task 4

# Problem 1: Mobile Recharge Validation System

def validate_mobile_recharge(amount):
  """
  This function checks whether the recharge amount is valid.
  conditions:
  1. Minimum recharge amount must be ₹50
  2. Only specific plans are allowed
  """
  valid_plans = [199, 299, 399, 599]

  # Check minimum recharge amount
  if amount < 50:
      return "Minimum recharge amount is ₹50"

  # Check if selected plan exists
  if amount not in valid_plans:
      return "Selected recharge plan is not available"

  # If all conditions pass
  return "Recharge Approved"

# Retry system using while loop
while True:
    user_amount = int(input("Enter recharge amount: ₹"))
    status = validate_mobile_recharge(user_amount)
    if status == "Recharge Approved":
        print("Recharge Completed Successfully")
        break
    else:
        print("Recharge Failed:", status)
        print("Please try again.\n")
```

```python
# Problem 2: Inventory Reorder Alert System

def inventory_status(stock_details):
  """
  This function checks each product's stock level.
  If stock is less than 15 -> Reorder Alert
  Otherwise -> Stock OK
  """
  for product, quantity in stock_details.items():
      if quantity < 15:
          print(f"{product} → Reorder Alert")
      else:
          print(f"{product} → Stock OK")

# Sample inventory data
inventory_data = {
    "Milk": 12,
    "Bread": 25,
    "Butter": 8,
    "Cheese": 30
}
inventory_status(inventory_data)
```

```python
# Problem 3: Student Result Processing System

def calculate_result(marks_list):
    """
    This function calculates the average marks and return Pass if average >= 50 else Fail.
    """
    total_marks = 0

    # Loop to calculate total
    for mark in marks_list:
        total_marks += mark
    average_marks = total_marks / len(marks_list)
    if average_marks >= 50:
        return "Pass"
    else:
        return "Fail"

# Sample marks
student_marks = [60, 45, 70, 55]
result = calculate_result(student_marks)
print("Result:", result)
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def compute_fare(distance_km, peak_time):
    """
    Fare Calculation Rules:
    - Base fare = ₹50
    - ₹12 per km
    - 25% extra during peak hours
    """
    base_fare = 50
    per_km_rate = 12
    total_fare = base_fare + (distance_km * per_km_rate)

    # Apply peak hour surcharge
    if peak_time =="yes":
        total_fare += total_fare * 0.25
    return total_fare

# Retry loop
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    final_amount = compute_fare(distance, peak)
    print("Estimated Fare: ₹", final_amount)
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
# Problem 5: Employee Attendence Eligibility Checker

def attendance_eligibility(attendance_record):
    """
    This function calculates attendence percentage.
    If attendance >= 75% -> Eligible
    Otherwise -> Not Eligible
    """
    present_count = 0

    # Count present days
    for status in attendance_record:
        if status == "P":
            present_count += 1
    percentage = (present_count / len(attendance_record)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Sample attendance data
attendance_data = ["P", "P", "A", "P", "P"]
print("Attendance Status:", attendance_eligibility(attendance_data))

```

```python
# Problem 6: Password Strength Checker

def password_strength(password):
    """
    Password Rules:
    - Minimum length = 8
    - Must contain at least one digit
    - Must contain at least one special character(@#$)
    """
    has_digit = False
    has_special = False
    special_characters = "@#$"

    # Check minimum Length
    if len(password) < 8:
        return "Weak Password (Minimum length 8 required)"

    # Loop through characters
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_characters:
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Missing digit or special character)"

# Sample password
test_password = "abc123@x"
print(password_strength(test_password))
```
